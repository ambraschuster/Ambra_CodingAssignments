### Explanation of the export code

`summons_sample` is the dataframe we want to export

`.to_csv()` saves our dataframe to a .csv file. You are baseically exporting your python data out into a format that Excel, Google Sheets can read.

`index=False`
By default, on export pandas includes the DataFrame's index – the row numbers (0, 1, 2, 3...). We NEVER
need it 99.99 percent of the time. `index=False` simply tells pandas to leave that column out on export.

`encoding="utf-8"`
Tells Python which character encoding to use when writing the file. `UTF-8` is the most universal standard — it handles English plus accented characters, symbols, non-Latin scripts, etc. without garbling them. If your data has names like "García" or "Müller" and you skip this, you might get weird characters or errors depending on your system.

````python
# 3. What are the total number of summons broken down by year?

# Group the data by the YEAR column and count the number of summons per year,
# then reset the index to convert the result into a clean DataFrame
summons_by_year = (
    nypd_summons.groupby("YEAR")          # Group rows by year
    .size()                                # Count the number of rows per group
    .reset_index(name="COUNT")             # Convert index to column, name the count column "COUNT"
    .sort_values("YEAR", ascending=True)   # Sort by year in ascending order (2015 → 2025)
    .reset_index(drop=True)                # Reset index after sorting for clean output
)

# Display the result
summons_by_year
````

### Prompt 7

Provide the pandas code to calculate which borough hat the fewest offenses per 100,000 people. These are the most recent population numbers from the 2020 census. populations = {
    "BROOKLYN": 2736074,
    "QUEENS": 2405464,
    "MANHATTAN": 1694251,
    "BRONX": 1472654,
    "STATEN ISLAND": 495747
}
Please ask me clarifiying questions until you are sure to be able to provide a correct code.

````python
#7. Which borough had the fewest offenses per 100K people?

# Define the 2020 census population for each of the 5 boroughs
populations = {
    "BROOKLYN": 2736074,
    "QUEENS": 2405464,
    "MANHATTAN": 1694251,
    "BRONX": 1472654,
    "STATEN ISLAND": 495747
}

# Filter the dataset to only include rows from the 5 boroughs,
# excluding nulls and any entries not in the populations dictionary
nypd_summons_boroughs = nypd_summons[                        # Filter the original DataFrame
    nypd_summons["BORO"].isin(populations.keys())            # Keep only rows matching the 5 borough names
]

# Count the number of summons per borough,
# then reset the index to convert the result into a clean DataFrame
summons_by_borough = (
    nypd_summons_boroughs.groupby("BORO")                    # Group rows by borough
    .size()                                                   # Count the number of rows per group
    .reset_index(name="COUNT")                                # Convert index to column, name it "COUNT"
)

# Map the population for each borough from the populations dictionary
summons_by_borough["POPULATION"] = summons_by_borough["BORO"].map(populations)  # Add population column

# Calculate the number of summons per 100,000 people for each borough
summons_by_borough["SUMMONS_PER_100K"] = (
    summons_by_borough["COUNT"] /                            # Divide summons count by population
    summons_by_borough["POPULATION"] * 100000                # Multiply by 100,000 to normalize
).round(2)                                                   # Round to 2 decimal places for readability

# Sort by summons per 100,000 ascending — fewest offenses per capita first
summons_by_borough = (
    summons_by_borough
    .sort_values("SUMMONS_PER_100K", ascending=True)         # Sort ascending — fewest per capita first
    .reset_index(drop=True)                                   # Reset index after sorting for clean output
)

# Display the result
summons_by_borough
````

### Finding 7:

Staten Island had the fewest summons issued per 100,000 people between 2015 and 2025, with about 9,400 summons per 100,000 residents.
