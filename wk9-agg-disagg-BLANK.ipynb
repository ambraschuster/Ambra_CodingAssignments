# Data `Aggregation` &  `Disaggregation`


```Data aggregation``` is the process by which we take data across rows or down columns and express it in a summary statistical form. It's therefore also known as a `summary analysis`.

A ```summary analysis``` calculates measures of central tendency (mean, median, mode) or variation (range, standard deviation).

Think of it as a way to get a high-level snapshot of outcomes or trends for a broad population, such as an entire country (or state or city), or for an industry.

## ```aggregate()```

The ```aggregate()``` method is really powerful (and a time saver). It allows us to apply one or more aggregate functions to a DataFrame or a Series.

Note that ```.aggregate()``` has an alias of ```.agg()```. We can use them interchangeably, but ```agg()``` is far more commonly used.

## Single columns with multiple statistical summaries

# The problem with `aggregation` (and a solution):
```Aggregation``` can mask important differences between subgroups and does not capture the full diversity of experiences within a population or industry.

But `disaggregation`, the opposite of ```aggregation```, is what you need here. `Disaggregation` breaks down data into smaller, more specific subgroups based on characteristics like income levels, gender, race, or other relevant demographics.

```Disaggregated``` data is critical in surfacing inequities and disparities.

## Data disaggregation in Pandas

## Question #6 - using AI 
### Since 1980, is there any country not using oil as their main and most emission causing fuel?

### Prompt for Question #6 
Provide the pandas code to calculate if there has been any country not using oil as their main and most emission causing fuel since 1980

````python
# Group the data by country and find the fuel with the maximum emissions for each country
# 1. Group the filtered dataset by 'Country' and 'fuel'
# 2. Calculate the sum of emissions for each country and fuel type combination
# 3. Reshape the data using unstack() to create a DataFrame where each row represents a country
#    and each column represents a fuel type
# 4. Find the fuel type with the maximum emissions for each country using idxmax(axis=1)
country_max_emissions = df_em.groupby(['Country', 'fuel'])['emissions'].sum().unstack().idxmax(axis=1)

# Check if there is any country not using oil as their main emission-causing fuel
# 5. Compare country_max_emissions with 'Oil' and select the index (country names)
#    where the condition is true
not_using_oil = country_max_emissions[country_max_emissions != 'Oil'].index.tolist()

# 6. If there are countries not using oil as their main emission-causing fuel, print them
if not_using_oil:
    print(f"Countries not using oil as their main emission-causing fuel since 1980: {', '.join(not_using_oil)}")
# 7. Otherwise, print a message stating that all countries have used oil as their main emission-causing fuel since 1980
else:
    print("All countries have used oil as their main emission-causing fuel since 1980.")
````

### Question 7 - using AI
#### Of all countries not using oil as their main emission causing fuel, what were their top 3 fuels?

### Prompt for Question #7
Provide the pandas code to calculate what the top 3 fuels where in countries where oil was not their main emission causing fuel.

````python
# 1. Group the data by country and find the fuel with the maximum emissions for each country
country_max_emissions = df_em.groupby(['Country', 'fuel'])['emissions'].sum().unstack().idxmax(axis=1)

# 2. Get the list of countries not using oil as their main emission-causing fuel
not_using_oil = country_max_emissions[country_max_emissions != 'Oil'].index.tolist()

# 3. Filter the dataset to include only countries not using oil as their main fuel
df_not_using_oil = df_em[df_em['Country'].isin(not_using_oil)]

# 4. Group the filtered dataset by fuel and calculate the total emissions for each fuel
fuel_emissions = df_not_using_oil.groupby('fuel')['emissions'].sum()

# 5. Sort the fuel emissions in descending order and get the top 3 fuels
top_3_fuels = fuel_emissions.nlargest(3)

# 6. Print the top 3 fuels and their emissions
print("Top 3 fuels in countries not using oil as their main emission-causing fuel:")
for fuel, emissions in top_3_fuels.items():
    print(f"{fuel}: {emissions:.2f}")
````

### Response #7 Countries who did not use oil as their main emission-causing fuel, heavily relied on gas and coal.
